# PHYLOAD data loading demo

This notebook demonstrates how to initialise datasets and dataloaders for the PDE HDF5 files stored under `scratch_jz/data`.

If your data lives somewhere else, set the environment variables `PHYLOAD_DATA_ROOT` and/or `PHYLOAD_DATASET` before launching the notebook.

```python
from pathlib import Path
import os
import yaml

import numpy as np

import matplotlib.pyplot as plt

from phyload import init_datasets, init_dataloaders
```

```python
with open('/lustre/fswork/projects/rech/ywo/urv67cz/code/PHYLOAD/configs/main.yaml') as f:
    cfg = yaml.safe_load(f)

cfg['data']['dataset_name'] = 'turbulent_radiative_layer_3D'
cfg['data']['use_normalization'] = False
cfg['loaders']['batch_size'] = 2
cfg['data']['dx'] = None
cfg['data']['window_stride'] = 1
cfg['data']['num_steps'] = None
cfg['loaders']['num_workers'] = 0
cfg['loaders']['prefetch_factor'] = 2
cfg['loaders']['persistent_workers'] = False

datasets = init_datasets(cfg['data'])
train_loader, val_loader, test_loader = init_dataloaders(datasets, cfg['loaders'])
```

```python
first_batch = next(iter(train_loader))
shapes = {
    'trajectory': first_batch['trajectory'].shape,
    'ic': first_batch.get('ic').shape if 'ic' in first_batch else None,
    'space_grid': first_batch['space_grid'].shape,
    'time_grid': first_batch['time_grid'].shape,
}
shapes
```

```python
import time

it = iter(train_loader)
times = []

for i in range(len(train_loader)):
    t0 = time.perf_counter()
    batch = next(it)                 # just fetching the batch
    t1 = time.perf_counter()
    times.append(t1 - t0)
    print(f"Batch {i}: load {t1 - t0:.4f}s")
    del batch
    if i == 20:
        break

print(f"Avg load time: {sum(times)/len(times):.4f}s")
```

```python
def save_trajectory_pdf(
    batch,
    dataset_name,
    *,
    channel,
    channel_names=None,
    time_steps=None,
    batch_idx=0,
    output_root=Path('visualizations'),
    figsize=(4, 4),
    cmap='viridis'
):
    if isinstance(channel, str):
        if channel_names is None:
            raise ValueError('channel_names must be provided when channel is a string.')
        try:
            channel_idx = channel_names.index(channel)
        except ValueError as exc:
            raise KeyError(f"Unknown channel '{channel}'. Available: {channel_names}") from exc
        channel_label = channel
    else:
        channel_idx = int(channel)
        if channel_names and 0 <= channel_idx < len(channel_names):
            channel_label = channel_names[channel_idx]
        else:
            channel_label = f'channel_{channel_idx}'

    trajectory = batch['trajectory'][batch_idx].detach().cpu().numpy()
    if channel_idx < 0 or channel_idx >= trajectory.shape[-1]:
        raise IndexError(f'Channel index {channel_idx} out of range for trajectory last dimension {trajectory.shape[-1]}.')

    spatial_dims = trajectory.ndim - 2  # remove time and component axes
    component_data = trajectory[..., channel_idx]
    time_values = batch['time_grid'][batch_idx].detach().cpu().numpy()
    total_steps = trajectory.shape[0]

    if time_steps is None:
        frame_indices = list(range(min(total_steps, 6)))
    else:
        candidate = []
        for step in time_steps:
            if isinstance(step, int) or (hasattr(step, 'is_integer') and float(step).is_integer()):
                idx = int(step)
            else:
                step_val = float(step)
                idx = int(abs(time_values - step_val).argmin())
            if idx < 0 or idx >= total_steps:
                raise ValueError(f"Requested time step {step} is out of range for this trajectory.")
            candidate.append(idx)
        frame_indices = []
        seen = set()
        for idx in candidate:
            if idx not in seen:
                frame_indices.append(idx)
                seen.add(idx)
        if not frame_indices:
            raise ValueError('No valid time steps selected for plotting.')

    data = component_data[frame_indices]
    space_grid = batch['space_grid'][batch_idx].detach().cpu().numpy()
    grid_selection = space_grid[frame_indices]
    selected_times = time_values[frame_indices]

    file_entries = batch['index']['file']
    file_name = file_entries[batch_idx] if isinstance(file_entries, (list, tuple)) else file_entries
    traj_entries = batch['index']['trajectory']
    traj_id = int(traj_entries[batch_idx]) if hasattr(traj_entries, '__len__') else int(traj_entries)
    window_entries = batch['index'].get('window_start')
    window_start = int(window_entries[batch_idx]) if window_entries is not None else 0

    output_dir = Path(output_root) / dataset_name / channel_label
    output_dir.mkdir(parents=True, exist_ok=True)
    pdf_path = output_dir / f"{Path(file_name).stem}_traj{traj_id}_start{window_start}.pdf"

    fig, axes = plt.subplots(1, len(frame_indices), figsize=(len(frame_indices) * figsize[0], figsize[1]), squeeze=False)
    axes = axes[0]
    for pos, (ax, frame_idx, t_val) in enumerate(zip(axes, frame_indices, selected_times)):
        frame_data = data[pos]
        grid_frame = grid_selection[pos]

        if spatial_dims == 1:
            x = grid_frame[0]
            ax.plot(x, frame_data)
            ax.set_xlabel('x')
            ax.set_ylabel('value')
        elif spatial_dims == 2:
            x = grid_frame[0]
            y = grid_frame[1]
            extent = [x.min(), x.max(), y.min(), y.max()]
            im = ax.imshow(frame_data, extent=extent, origin='lower', aspect='auto', cmap=cmap)
            fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
            ax.set_xlabel('x')
            ax.set_ylabel('y')
        elif spatial_dims == 3:
            z_mid = frame_data.shape[0] // 2
            x = grid_frame[0][z_mid]
            y = grid_frame[1][z_mid]
            extent = [x.min(), x.max(), y.min(), y.max()]
            im = ax.imshow(frame_data[z_mid], extent=extent, origin='lower', aspect='auto', cmap=cmap)
            fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
            ax.set_xlabel('x')
            ax.set_ylabel('y')
            ax.set_title(f'z slice {z_mid}')
        else:
            ax.text(0.5, 0.5, 'Unsupported spatial dim', ha='center', va='center')
            ax.axis('off')
            continue
        ax.set_title(f"t={t_val:.2f} (idx={frame_idx})")

    fig.suptitle(f"{dataset_name} | {channel_label} | file {file_name} | traj {traj_id}")
    fig.tight_layout()
    fig.savefig(pdf_path, format='pdf')
    plt.show()
    plt.close(fig)
    return pdf_path
```

```python
channel_names = datasets['train'].channel_names
pdf_path = save_trajectory_pdf(
    first_batch,
    cfg['data']['dataset_name'],
    channel=channel_names[0],
    channel_names=channel_names,
    time_steps=[0, 25, 50, 75, 100],
    batch_idx=0,
    figsize = (4, 4),
    cmap = 'RdBu'
)
pdf_path
```
