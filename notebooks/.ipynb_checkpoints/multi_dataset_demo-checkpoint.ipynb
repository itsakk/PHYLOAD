# Multi-dataset loading demo
Example illustrating how to combine several PHYLOAD datasets into a single training loader.

```python
import yaml
from phyload import init_datasets, init_dataloaders

with open('/lustre/fswork/projects/rech/ywo/urv67cz/code/PHYLOAD/configs/main_multi.yaml') as f:
    cfg = yaml.safe_load(f)

datasets = init_datasets(cfg["data"])
train_loader, val_loader, test_loader = init_dataloaders(datasets, cfg["loaders"])

example_batch = next(iter(train_loader))
example_keys = list(example_batch.keys())
example_keys
```

```python
# Inspect padded shapes in mixed mode

for example_batch in train_loader:
    print('trajectory shape:', example_batch['trajectory'].shape)
    print('dataset ids:', example_batch['dataset'])
```
